# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Importing the dataset

```python
Column_Labels = ['Sepal length', 'Sepal width', 'Petal length', ' Petal width', 'Species']
# Load the data for iris
df =pd.read_csv('iris.data', names = Column_Labels )
```

# Basic statistical analysis

```python
#Describing dataset for setosa species
df[df['Species']=='Iris-setosa'].describe()
```

# Visualising the whole dataset 

```python
sns.pairplot(df, hue = 'Species')
```

# Seperating the features and dependent variables


```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

# Splitting the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

# Training the dataset by Support vector machine algorithm

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, Y_train)
```

# Predicting the dataset 

```python
Y_pred = classifier.predict(X_test)
```

# Calculating the accuracy of the prediction

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(accuracy_score(Y_test, Y_pred))
```
